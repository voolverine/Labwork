```python
import pandas as pd
import matplotlib
from matplotlib import pylab as plt
import numpy as np
from IPython.display import display
import scipy.stats as st

def phi(X):
    return map(lambda x: x ** 2, X)

def expected_value_evaluation(n, a, b, teoretical_varience):
    
    eps = np.random.random(n)    
    X = eps * (b - a) + a
    Y = phi(X)

    # Practice expected value calculation
    practice_ev = sum(Y) / float(n)
    
    # Practice varience calculation
    practice_varience = 0.0
    for Yi in Y:
        practice_varience += (Yi - practice_ev) ** 2 / float(n - 1)
    
    
    def get_interval_of_confidence(varience, confidence_level):
        eps = (varience / n) ** 0.5 * st.norm.ppf((1 + confidence_level) / 2.0)
        return (practice_ev - eps, practice_ev + eps)
    
    # Graphic of interval of confidence by confidence_level
    Expected_Cinterval_x = matplotlib.mlab.frange(0.01, 0.99, 0.01)
    Expected_Cinterval_y = []
    for i in Expected_Cinterval_x:
        interval = get_interval_of_confidence(practice_varience, i)
        Expected_Cinterval_y.append(interval[1] - interval[0])
        
    plt.plot(Expected_Cinterval_x, Expected_Cinterval_y)
    
    
    # Graphic of interval of confidence by confidence_level with varience number known
    Expected_Cinterval_TV_x = matplotlib.mlab.frange(0.01, 0.99, 0.01)
    Expected_Cinterval_TV_y = []
    for i in Expected_Cinterval_TV_x:
        interval = get_interval_of_confidence(teoretical_varience, i)
        Expected_Cinterval_TV_y.append(interval[1] - interval[0])  
        
    plt.plot(Expected_Cinterval_TV_x, Expected_Cinterval_TV_y, 'r')
    plt.show()
    return
          

def varience_evaluation(n, a, b, teoretical_ev):
    eps = np.random.random(n)    
    X = eps * (b - a) + a
    Y = phi(X)

    # Practice expected value calculation
    practice_ev = sum(Y) / float(n)
    
    # Practice varience calculation
    practice_varience = 0.0
    for Yi in Y:
        practice_varience += (Yi - practice_ev) ** 2 / float(n - 1)
    
    def get_interval_of_confidence(varience, confidence_level):
        eps = varience * ((0.8 * n + 1.2) / (n * (n - 1))) ** 0.5 \
                * st.norm.ppf((1 + confidence_level) / 2.0)
        return (practice_varience - eps, practice_varience + eps)
    
    # Graphic of interval of confidence by confidence_level
    Expected_Cinterval_x = matplotlib.mlab.frange(0.01, 0.99, 0.01)
    Expected_Cinterval_y = []
    for i in Expected_Cinterval_x:
        interval = get_interval_of_confidence(practice_varience, i)
        Expected_Cinterval_y.append(interval[1] - interval[0])
        
    plt.plot(Expected_Cinterval_x, Expected_Cinterval_y)
    
    # Varience with expected number known
    practice_varience_known = 0.0
    for Yi in Y:
        practice_varience_known += (Yi - teoretical_ev) ** 2 / float(n - 1)

    # Graphic of interval of confidence by confidence_level with expected number known
    Expected_Cinterval_TV_x = matplotlib.mlab.frange(0.01, 0.99, 0.01)
    Expected_Cinterval_TV_y = []
    for i in Expected_Cinterval_TV_x:
        interval = get_interval_of_confidence(practice_varience_known, i)
        Expected_Cinterval_TV_y.append(interval[1] - interval[0])  
        
    plt.plot(Expected_Cinterval_TV_x, Expected_Cinterval_TV_y, 'r')
    plt.show()
    
    return

def draw_CI_length_by_series_length(n, a, b, varience):
    eps = np.random.random(n)    
    X = eps * (b - a) + a
    Y = phi(X)

    # Returns length of confidence interval of expected value(EV) with confidence level
    def get_EVCI_length(n, confidence_level):
        eps = (varience / n) ** 0.5 * st.norm.ppf((1 + confidence_level) / 2.0)
        # eps = (varience / n) ** 0.5 * (2.0/float(n-1)) * st.norm.ppf((1 + confidence_level) / 2)
        return 2 * eps
    
    # Returns length of confidence interval of varience(V) with confidence level
    def get_VCI_length(n, confidence_level):
        eps = varience * ((0.8 * n + 1.2) / (n * (n - 1))) ** 0.5 \
                * st.norm.ppf((1 + confidence_level) / 2.0)
        #eps = varience * (2.0/float(n-1)) * st.norm.ppf((1 + confidence_level) / 2)
        return 2 * eps
    
    # Assume confidence level = 0.9
    
    confidence_level = 0.9
    
    
    # EVCI Graphic drawing
    EVCI_x = matplotlib.mlab.frange(2, n, 1)
    EVCI_y = []
    for i in EVCI_x:
        EVCI_y.append(get_EVCI_length(i, confidence_level))
        
    plt.plot(EVCI_x, EVCI_y, 'b')
        
        
    # VCI Graphic drawing 
    VCI_x = matplotlib.mlab.frange(2, n, 1)
    VCI_y = []
    for i in VCI_x:
        VCI_y.append(get_VCI_length(i, confidence_level))
    plt.plot(VCI_x, VCI_y, 'r')
    
    plt.show()

    
def main():

    n = int(raw_input("n = "))
    expected_value_evaluation(n, 0, 6, 115.2)   
    varience_evaluation(n, 0, 6, 12.0)    
    draw_CI_length_by_series_length(150, 0, 6, 115.2)
    
if __name__ == "__main__":
    main()
```
